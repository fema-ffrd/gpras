# Visualizing toy data feature space


```python
import matplotlib.pyplot as plt

from gpras.utils.preprocessing import ToyTrainingData
```

```python
toy_data = ToyTrainingData(resolution=100)
```

```python
# Plot the toy elevation
fig, ax = plt.subplots()
ax.imshow(toy_data.elevation_grid, origin="lower")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
# Plot some feature planes
fig, ax = plt.subplots()
x = toy_data.inputs[:, 0]
y = toy_data.inputs[:, 1]
z = toy_data.inputs[:, 2]

ax.scatter(x, y, c=z)
ax.set_aspect('equal')
ax.set_xlabel("X")
ax.set_ylabel("Y")
```
